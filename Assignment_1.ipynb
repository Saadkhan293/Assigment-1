```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
gr = np.column_stack([[x, y] for x in range(-15, 16, 1) for y in range(-15, 16, 1)])

plt.figure(figsize=(10, 10), facecolor="w")
plt.scatter(gr[0], gr[1], s=36, edgecolor="none")
# Set axis limits
plt.grid(True)
plt.axis("equal")
plt.xlim(-30,30)
plt.ylim(-30,30)
plt.title("First Grid")
```

```python
gr = np.column_stack([[x, y] for x in range(-15, 16, 1) for y in range(-15, 16, 1)])

for i in range(len(gr[0])):
    gr[0][i] += gr[1][i]
        
plt.figure(figsize=(10, 10), facecolor="w")
plt.scatter(gr[0], gr[1], s=36, edgecolor="none")
plt.grid(True)
plt.axis("equal")
plt.xlim(-30,30)
plt.ylim(-30,30)
plt.title("Shear Transformed Gird With Loops")
```

```python
gr = np.column_stack([[x, y] for x in range(-15, 16, 1) for y in range(-15, 16, 1)])
gr[0] = gr[0]+gr[1]

plt.figure(figsize=(10, 10), facecolor="w")
plt.scatter(grid[0], gr[1], s=36, edgecolor="none")
plt.grid(True)
plt.axis("equal")
plt.xlim(-30,30)
plt.ylim(-30,30)
plt.title("Shear Transformed Gird Without any Loops")
```
